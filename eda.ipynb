```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("Superstore.csv",encoding="latin1")
```

```python
df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    format='mixed',
    dayfirst=True
)
```

```python
df['Ship Date'] = pd.to_datetime(
    df['Ship Date'],
    format='mixed',
    dayfirst=True
)
```

```python
df.dtypes
```

```python
df.columns
```

```python

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#freature Engineering
```

```python
df['year']=df['order_date'].dt.year
df['month_name']=df['order_date'].dt.month_name()
df['month']=df['order_date'].dt.month
df['quarter']=df['order_date'].dt.quarter
df["Day"] = df["order_date"].dt.day_name()
```

```python
df.head()
```

```python
df['profit_margin']=((df['profit']/df['sales'])* 100).round(2)
```

```python
df["shipping_days"] = (
    df["ship_date"] - df["order_date"]
).dt.days
```

```python
df.head()
```

```python
df["order_value"] = pd.cut(
    df["sales"],
    bins=[0, 100, 500, float("inf")],
    labels=["Low", "Medium", "High"]
)
```

```python
df.head()
```

```python
#EDA
```

```python
total_sales = df["sales"].sum()
print(f"Total Sales: {total_sales:,.2f}")
```

```python
total_profit = df["profit"].sum()
print(f"Total Profit: {total_profit:,.2f}")
```

```python
total_orders = df["order_id"].nunique()
print(f"Total Orders: {total_orders}")
```

```python
average_order_value = total_sales / total_orders
print(f"Average Order Value: {average_order_value:.2f}")
```

```python
sales_region = df.groupby("region")["sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sales_region.plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python

sales_category = df.groupby("category")["sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sales_category.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
df.columns = df.columns.str.replace('-','_')
df.columns
sales_sub_category = df.groupby("sub_category")["sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sales_sub_category.plot(kind="bar")
plt.title("Sales by Sub Category")
plt.xlabel("Sub Category")
plt.ylabel("Sales")
plt.show()
```

```python
profit_category = df.groupby("category")["profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
profit_category.plot(kind="bar")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
monthly_sales = df.groupby("month")["sales"].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
monthly_profit = df.groupby("month")["profit"].sum()

plt.figure(figsize=(10,5))
monthly_profit.plot(marker="o")
plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
top_products = (
    df.groupby("product_name")["sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))
top_products.plot(kind="bar")
plt.title("Top 10 Products by Sales")
plt.ylabel("Sales")
plt.show()
```

```python
top_customers = (
    df.groupby("customer_name")["sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))
top_customers.plot(kind="bar")
plt.title("Top 10 Customers")
plt.ylabel("Sales")
plt.show()
```

```python
top_cities = (
    df.groupby("city")["sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))
top_cities.plot(kind="bar")
plt.title("Top Cities by Sales")
plt.ylabel("Sales")
plt.show()
```

```python
top_states = (
    df.groupby("state")["profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))
top_states.plot(kind="bar")
plt.title("Top States by Profit")
plt.ylabel("Profit")
plt.show()
```

```python
segment_sales = df.groupby("segment")["sales"].sum()

plt.figure(figsize=(7,5))
segment_sales.plot(kind="bar")
plt.title("Sales by Segment")
plt.ylabel("Sales")
plt.show()
```

```python
ship_mode = df.groupby("ship_mode")["profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
ship_mode.plot(kind="bar")
plt.title("Profit by Ship Mode")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["discount"], df["profit"])
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["number"])

plt.figure(figsize=(12,10))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Q1 = df["sales"].quantile(0.25)
Q3 = df["sales"].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df["sales"] < Q1 - 1.5 * IQR) |
    (df["sales"] > Q3 + 1.5 * IQR)
]

print(outliers.shape)
outliers.head()
```

```python
customer_segment = (
    df.groupby("customer_name")
      .agg({
          "sales":"sum",
          "profit":"sum",
          "order_id":"nunique"
      })
      .sort_values(by="sales", ascending=False)
)

customer_segment.head()
```

# ✅ Conclusion

## Key Findings

- Sales and profit vary significantly across regions.
- Technology products generate strong profitability.
- Furniture contributes high sales but comparatively lower profit.
- A small percentage of customers contribute a significant share of revenue.
- Higher discounts often reduce profitability.
- Seasonal trends are visible in monthly sales performance.
